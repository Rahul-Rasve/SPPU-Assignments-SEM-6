```python
import nltk
nltk.download('punkt')
```

```python
from nltk import tokenize
from nltk.tokenize import sent_tokenize
```

```python
#read text
text = "Good Day Everyone, How are you all today? Its Fun learning Data Analytics. Hope you all are practicing well"
text
```

```python
tokens = sent_tokenize(text)
tokens
```

```python
from nltk.tokenize import word_tokenize
word_tokens = word_tokenize(text)
word_tokens
```

```python
#calc frquency of words
from nltk.probability import FreqDist
fdist = FreqDist(word_tokens)
fdist
```

```python
fdist.most_common(3) #print 3 most common
```

```python
#plots
import matplotlib.pyplot as plt
```

```python
fdist.plot(30, cumulative=False)
plt.show()
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
#get stop words in english
stop_words = set(stopwords.words("english"))
```

```python
filtered_sent = []
for word in word_tokens:
  if word not in stop_words:
    filtered_sent.append(word)

print('Tokenized Sentence : ', word_tokens)
print('Filtered Sentence : ', filtered_sent)
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
stemmer_words = [] #lowercase words
for word in filtered_sent:
  stemmer_words.append(ps.stem(word)) #convert to lowercase

print('Filtered Sentence : ', filtered_sent)
print('Stemmed Sentence : ', stemmer_words)
```
