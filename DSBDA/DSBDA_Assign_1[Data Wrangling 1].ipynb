```python
#Data Formatting and Normalization
```

```python
#replace null values with 'Unknown'
df["Cabin"] = df["Cabin"].replace(to_replace=np.nan, value="UnKnown")
df
```

```python
#fill the null values with a closer value
df["Age"] = df["Age"].interpolate()
df
```

```python
#fill the null values with 
df["Embarked"].fillna(method="ffill", inplace=True)
df
```

```python
#removed all null values
df.isnull().sum()
```

```python
#changeing datatypes
df["Age"] = df["Age"].astype('int64')
df.dtypes
```

```python
#Changing Categorical Variables into Quantitative Variables
```

```python
#create dummy variables (1 or 0)
quan_data = pd.get_dummies(df.Embarked, prefix = 'Embarked')
quan_data
```

```python
#join the original data frame with new "quan_data"
df = df.join(quan_data)
df
```

```python
#categorical data is converted to quantitative, so drop categorical col
df.drop(["Embarked"], axis=1, inplace=True)
df
```

```python
#'sex' is also categorical (Male or Female) col
quan_sex = pd.get_dummies(df.Sex, prefix="Sex")
quan_sex
```

```python
#join it to the original data frame
df = df.join(quan_sex)
df
```
