```python
x = df.iloc[:, 0:4].values #'0:4 means coluns 0 to 4 Excluding 4 OR [0,4)'
x
```

```python
y = df.iloc[:, 4].values
y
```

```python
#label encoding
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
y = le.fit_transform(y) #this will give each species a number
y
```

```python
#split data into test and train
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=30)
```

```python
x_train
```

```python
#initializing Naive Bayes Model and training it
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train, y_train)
```

```python
#prediction
prediction = model.predict(x_test)
prediction
```

```python
y_test
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, prediction)
cm
```

```python
#accuracy
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, prediction)*100
acc
```

```python
#precision
from sklearn.metrics import precision_score
prc = precision_score(y_test, prediction, average='micro')*100
prc
```

```python
#recall
from sklearn.metrics import recall_score
recall = recall_score(y_test, prediction, average='micro')*100
recall
```
