```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/Dataset/Social_Network_Ads.csv')
df
```

```python
#select the 2nd and 3rd columns (0-indexed)
x = df.iloc[:, [2,3]].values #here ':' means all rows and '[2,3]' means col 2 & 3
x
```

```python
#select the column 4
y = df.iloc[:, 4].values
y
```

```python
#split test and train data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.25, random_state=40)
```

```python
print('x_train : ', x_train.shape, '  y_train : ', y_train.shape)
```

```python
print('x_test : ', x_test.shape, '  y_test : ', y_test.shape)
```

```python
#data standardization
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc_x = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.fit_transform(x_test)
```

```python
x_train
```

```python
#initialize logistic regression model and trainig the model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train, y_train)
```

```python
#prediction
y_pred = classifier.predict(x_test)
y_pred
```

```python
y_test
```

```python
#confusion matrix
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
#true positive : Actually True, Predicted True
tp = cm[0, [0]]
print('True Positive : ', tp)
```

```python
#false positive : Actually False, Predicted True
fp = cm[0, [1]]
print('False Positive : ', fp)
```

```python
#false negative : Actually True, Predicted False
fn = cm[1, [0]]
print('False Negative : ', fn)
```

```python
#true negative : Actually False, Predicted False
tn = cm[1, [1]]
print('True Negative : ', tn)
```

```python
#total of all values
total = (tp + fp + fn + tn)
```

```python
#accuracy
accuracy = ((tp + tn) / total)*100 #for percent
print('Accuracy : ', accuracy)
```

```python
#precision
precision = (tp / (tp + fp))*100 #for percent
print('Precision : ', precision)
```

```python
#recall
recall = (tp / (tp + fn))*100
print('Recall/Sensitivity : ', recall)
```

```python
#specificity
specificity = (tn / (tn + fp))*100
print("Specificity : ", specificity)
```
