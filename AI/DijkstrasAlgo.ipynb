```python
# pronounced as -> dike struh ['j' is silent]

#define graph
graph = {
  'A':{'B' :2, 'C' :4},  #A is connected to B with distance 2 & C with distance 4
  'B':{'A' :2, 'C' :3, 'D' :8},
  'C':{'A' :4, 'B' :3, 'E' :5, 'D' :2},
  'D':{'B' :8, 'C' :2, 'E' :11, 'F' :22},
  'E':{'C' :5, 'D' :11, 'F' :1},
  'F':{'D' :22, 'E' :1}
}
```

```python
from numpy import Inf as max
from heapq import heapify, heappush, heappop

#keep track of visited nodes
visited = []

#keep track of every node's cost and predecessor/parent
node_data = {
  'A': {'cost': max, 'pred': []},
  'B': {'cost': max, 'pred': []},
  'C': {'cost': max, 'pred': []},
  'D': {'cost': max, 'pred': []},
  'E': {'cost': max, 'pred': []},
  'F': {'cost': max, 'pred': []},
}

#define dijkstra's algorithm
def dijkstra(graph, start, end):
  #initialize starting node
  node_data[start]['cost'] = 0

  #keep  track of minimum value
  current_min = start

  #iteration count
  i = 0

  #we iterate over the graph for (n-1) times since MST will have (n-1) edges
  while i != (len(graph)-1):
    if current_min not in visited:
      visited.append(current_min)
      min_heap = []

      #traverse the neighbours of current node
      for j in graph[current_min]:
        if j not in visited:

          # find the cost of current node from start
          cost = node_data[current_min]['cost'] + graph[current_min][j] #cost of parent + weight of edge

          #compare costs
          if cost < node_data[j]['cost']:
            node_data[j]['cost'] = cost #for a non_visited node, it will change th cost from infinity to 'cost'
            #add the predecessors of parent to the list
            node_data[j]['pred'] = node_data[current_min]['pred'] + list(current_min)

          #push the adjacent nodes onto the heap
          heappush(min_heap, (node_data[j]['cost'], j))

      heapify(min_heap) #heapify to get the min element on top
      current_min = min_heap[0][1]
    
    #increment counter
    i += 1
  
  #print the minimum cost & path
  print("Minimum Cost : " + str(node_data[end]['cost']))
  print("Shortest Path : " + str(node_data[end]['pred'] + list(end)))
```

```python
dijkstra(graph, 'A', 'F')
```
