```python
#define the required data
job = ['A', 'B', 'C', 'D', 'E']
deadline = [2, 1, 2, 1, 3]
profit = [100, 19, 27, 25, 15]

#combined
jobsNprofit = list(zip(job, deadline, profit))
print(jobsNprofit)
```

```python
#sort the list in descending order
jobsNprofit = sorted(jobsNprofit, key = lambda x: x[2], reverse = True)
print(jobsNprofit)
```

```python
def jobSchedule(jobsNprofit):
  #jobs scheduled
  result = ['null' for _ in range(len(jobsNprofit))]

  #define time line
  slot = [0 for _ in range(len(jobsNprofit))]

  #store total profit
  profit = 0

  #job scheduling algo loop
  for i in range(len(jobsNprofit)):
    job = jobsNprofit[i] #a tuple => (job , deadline, profit)

    #check the deadline on job
    for j in range(job[1], 0, -1): #go from job[1] which is deadline to 0 by decrementing by 1
      # check if slot is available
      if slot[j] == 0:
        #add job to result
        result[j] = job[0]
        profit += job[2]
        slot[j] = 1
        break
  print("Jobs Scheduled : ", result[1:])
  print("Total Profit : ", profit)
```

```python
jobSchedule(jobsNprofit)
```
