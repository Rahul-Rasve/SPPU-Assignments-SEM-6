```python
#graph
graph = {
    "A": [("B", 2), ("E", 3)],
    "B": [("C", 1), ("G", 9)],
    "C": None,
    "D": [("G", 1)],
    "E": [("D", 6)],
    "G": None
}

#heuristic values
def getHue(n):
  heu = {
      "A": 11,
      "B": 6,
      "C": 99,
      "D": 1,
      "E": 7,
      "G": 0
  }
  return heu[n]

#retreive adjacent nodes
def getAdjacent(vertex):
  if vertex in graph:
    return graph[vertex]
  else:
    return None
```

```python
#A* function
def astar(start, end):
  openSet = set(start)
  closeSet = set()
  g = {} #g(n)
  parent = {}

  #dist of starNode from itself is 0
  g[start] = 0
  parent[start] = start

  while len(openSet) > 0:
    n = None

    #get the least f(n)
    for vertex in openSet:
      if n==None or g[vertex] + getHue(vertex) < g[n] + getHue(n):
        n = vertex

    if n == end or graph[n] == None:
      pass
    else:
      for (m, wt) in getAdjacent(n):
        if m not in openSet and m not in closeSet:
          openSet.add(m)
          parent[m] = n
          g[m] = g[n] + wt

        #select smallest distance
        else:
          if g[m] > g[n] + wt:
            g[m] = g[n] + wt
            #change parent
            parent[m] = n

            #if m in closeSet, remove and add to open
            if m in closeSet:
              closeSet.remove[m]
              openSet.add(m)

    if n==None :
      print("Path does not Exist")
      return None

    #if current node is endNode, reconstruct from start to end
    if n==end:
      path = []

      while parent[n] != n:
        path.append(n)
        n = parent[n]

      #sdd atrat node
      path.append(start)
      path.reverse()

      return path

    openSet.remove(n)
    closeSet.add(n)
  
  print("Path does not exist")
  return None
```

```python
print('Path -> ', astar("A", "G"))
```
