```python
#define the is_safe function
def is_safe(row, col, board):
  #check negative diagonal
  irow, icol = row, col
  while irow>=0 and icol>=0: #from bottom right to top left
    if board[irow][icol] == 'Q':
      return False
    irow -= 1
    icol -= 1

  #check negative diagonal
  irow, icol = row, col
  while irow<len(board) and icol>=0: #from top right to bottom left
    if board[irow][icol] == 'Q':
      return False
    irow += 1
    icol -= 1

  #check the row
  irow, icol = row, col
  while icol >= 0: #from right to left
    if board[irow][icol] == 'Q':
      return False
    icol -= 1

  return True
```

```python
#define function to place the queen
def place_queens(size, result, board, col):
  #end column
  if col == size:
    result.append([''.join(row) for row in board]) #for each row it joins all the elements into a string
    return 

  for row in range(size):
    if is_safe(row, col, board):
      board[row][col] = 'Q'
      place_queens(size, result, board, col+1)
      board[row][col] = '.' #add '.' to rest of the row
```

```python
#problem solution
def solve(size):
  result = []
  board = [['.' for _ in range(size)] for _ in range(size)]
  print("Initial Board -> \n", board, "\n")
  place_queens(size, result, board, 0) #pass the start col
  return result
```

```python
N = int(input("Size of Board : "))
result_board = solve(N)

#print the result_board
for step in result_board:
  for row in step:
    print(row)
  print("\n<------------>\n")

print("\nTotal Solutions : ", len(result_board))
```
