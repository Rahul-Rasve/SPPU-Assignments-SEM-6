```python
#adjacency matrix
graph = [[0, 19, 5, 0, 0],
        [19, 0, 5, 9, 2],
        [5, 5, 0, 1, 6],
        [0, 9, 1, 0, 1],
        [0, 2, 6, 1, 0]]
```

```python
visited = []

#define prims algorithm
def prims(graph, N):
  visited.append(0) #starting node

  while len(visited) != N:
    a = 0
    b = 0
    min = 1000

    for node in range(N):
      if node in visited: #if node is visited
        for vertex in range(N):
          #if vertex is not visited and there exists an edge betw them
          if vertex not in visited and graph[node][vertex]:
            if min > graph[node][vertex]:
              min = graph[node][vertex]
              a = node
              b = vertex
    print(str(a) + "-" + str(b) + ":" + str(graph[a][b]))
    visited.append(b) #mark the vertex as visited
```

```python
prims(graph, 5)
```
