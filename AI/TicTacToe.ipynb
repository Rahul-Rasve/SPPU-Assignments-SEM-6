```python
#define the print function
def printBoard(xState, oState):
  zero  = 'X' if xState[0] else ('O' if oState[0] else 0)
  one   = 'X' if xState[1] else ('O' if oState[1] else 1)
  two   = 'X' if xState[2] else ('O' if oState[2] else 2)
  three = 'X' if xState[3] else ('O' if oState[3] else 3)
  four  = 'X' if xState[4] else ('O' if oState[4] else 4)
  five  = 'X' if xState[5] else ('O' if oState[5] else 5)
  six   = 'X' if xState[6] else ('O' if oState[6] else 6)
  seven = 'X' if xState[7] else ('O' if oState[7] else 7)
  eight = 'X' if xState[8] else ('O' if oState[8] else 8)

  print(f" {zero} | {one} | {two} ")
  print("---|---|---")
  print(f" {three} | {four} | {five} ")
  print("---|---|---")
  print(f" {six} | {seven} | {eight} ")
```

```python
#define the winning conditions
def checkWin(xState, oState):
  wins = [[0,1,2], [3,4,5], [6,7,8], [0,3,6], [1,4,7], [2,5,8], [0,4,8], [2,4,6]]

  for win in wins:
    if (xState[win[0]] + xState[win[1]] + xState[win[2]]) == 3:
      print("\nX won the game!")
      return 1
    elif (oState[win[0]] + oState[win[1]] + oState[win[2]]) == 3:
      print("\nO won the game!")
      return 0
    
  return -1
```

```python
#driver code
xState = [0]*9
oState = [0]*9

#initialize first turn
turn = 1

#track the count of moves done
moves = 0

print("Welcome to Tic-Tac-Toe")
print("\nEnter '10' to quit the game.\n")
while(True):
  printBoard(xState, oState)

  print("X's Turn!") if turn == 1 else print("O's Turn!")

  value = int(input("Please enter a value from Grid : "))
  
  if value in range(0,9):
    if turn == 1:
      xState[value] = 1
      moves += 1
    else:
      oState[value] = 1
      moves += 1
      
    roundResult = checkWin(xState, oState)

    if roundResult != -1:
      print("Match Over")
      break

    if moves == 9 and roundResult == -1:
      print("\nIt's a Tie!\n")
      moves = 0
      #reinitialize the states
      xState = [0]*9
      oState = [0]*9
      #set turn to 0 so it becomes X's turn, since (1-0)
      turn = 0

    turn = 1 - turn
  elif value == 10:
    print("\nThank You!\n")
    break
  else:
    print("\nEnter Valid Number!\n")
```
