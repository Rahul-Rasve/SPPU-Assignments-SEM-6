```python
#graph representation
graph = {
    "A": ["B", "C"],
    "B": ["D", "E"],
    "C": ["B", "F"],
    "D": [],
    "E": ["F"],
    "F": []
}
```

```python
#keep track of visitd nodes
visited = set() 

#store dfs output
output = []
```

```python
#dfs algo recursive function
def dfs(graph, visited, node):
  #check if the input node is visited : ??
  if node not in visited:
    output.append(node)
    visited.add(node)
    #iterate over its adjacent nodes
    for adjacent in graph[node]:
      dfs(graph, visited, adjacent)
```

```python
#execute method
dfs(graph, visited, "A")
print(output)
```
