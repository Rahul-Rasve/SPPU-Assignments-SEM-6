```python
#graph representation
graph = {
    "A": ["B", "C"],
    "B": ["D", "E"],
    "C": ["B", "F"],
    "D": [],
    "E": ["F"],
    "F": []
}
```

```python
#keep track of visited nodes
visited = set()
```

```python
import collections

#bfs algorithm
def bfs(graph, node):
  queue = collections.deque([node])
  while queue:
    vertex = queue.popleft() #pop from front of queue
    visited.add(vertex)
    for adjacent in graph[vertex]:
      if adjacent not in visited:
        queue.append(adjacent)
```

```python
#execution
bfs(graph, "A")
print(visited)
```
