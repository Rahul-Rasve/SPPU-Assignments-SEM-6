```python
#define list
arr = [16, 5, 4, 12, 10]
```

```python
#define selection sort function
def selectionSort(a):
  for i in range(len(a)):
    min_index = i   #for ascending, take min_index & max_index for descending
    for j in range(i+1, len(a)):
      if a[j] < a[min_index]:
        min_index = j
    #swap the elements of index (i, min_index)
    a[i], a[min_index] = a[min_index], a[i]
  print(a)
```

```python
selectionSort(arr)
```
