```python
# define graph
graph = [
    [1, 2, 3], #edges of node 0
    [0, 2],    #edges of node 1
    [0, 1, 3], #edges of node 2
    [0, 2]     #edges of node 3
]

colors = [0, 1, 2, 3]
```

```python
color_assigned = [-1]*len(graph)
size = len(graph)

def is_valid(vertex, color):
  for adj in graph[vertex]:
    if color_assigned[adj] == color: #if adjacent node has same color
      return False
  return True

def coloring_vertex(vertex):
  if vertex == size:
    return True
  
  for color in colors:
    if is_valid(vertex, color):
      color_assigned[vertex] = color

      if coloring_vertex(vertex+1):
        return True

  return False
```

```python
if coloring_vertex(0):
  print("Solution Found => ")
  for vertex in range(size):
    print(f"Vertex {vertex} : Color {color_assigned[vertex]}")
else:
  print("No solution")
```
